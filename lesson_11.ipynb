```python
import pandas as pd
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry') 
```

```python
drinks.head()
```

```python
# dropping particular column, axis=1
# if 'inplace=True' parameter is not specified, it will not aLter the the original DataFrame 
drinks.drop('continent', axis=1).head()
```

```python
# dropping multiple columns, axis=1
drinks.drop(['country','continent'], axis=1).head()
```

```python
# dropping particular row, axis=0
drinks.drop(2, axis=0).head()
```

```python
# dropping multiple rows, axis=0
drinks.drop([2,4,5], axis=0).head()
```

```python
# showing the mean value of every numeric column
drinks.mean()
# sames as drinks.mean(axis='index')
```

```python
# showing the mean value of every row with numeric value, axis=1
drinks.mean(axis=1).head(10)
# sames as drinks.mean(axis='columns')
```
