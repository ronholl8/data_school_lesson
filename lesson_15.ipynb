```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies.dtypes
```

```python
# use describe() method to explore Series
# count means total of non-null value
# unique means no of unique value
# top means the most common value
# freq means frequency of most common value
movies.genre.describe()
```

```python
# use of value_counts() method
movies.genre.value_counts()
```

```python
# convert the count into percentage with parameter normalize=True
movies.genre.value_counts(normalize=True)
```

```python
movies.genre.value_counts(normalize=True).head()
```

```python
# display list of unique values
movies.genre.unique()
```

```python
# display count of unique values
movies.genre.nunique()
```

```python
# display a tally shows the count of intersection between 'genre' and 'content_rating'
pd.crosstab(movies.genre, movies.content_rating)
```

```python
pd.crosstab(movies.content_rating, movies.genre)
```

```python
movies.duration.describe()
```

```python
movies.duration.mean()
```

```python
movies.duration.value_counts()
```

```python
movies.duration.value_counts(normalize=True)
```
