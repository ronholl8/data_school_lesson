```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
filter_col = ['title', 'genre']
movies[filter_col].head(10)
```

```python
type(movies[filter_col])
```

```python
movies[movies.duration >= 200]
```

```python
# multiple filter
# use of 'and'(&) condition
movies[(movies.duration >= 200) & (movies.genre == 'Drama')]
```

```python
# use of 'or'(|) condition
movies[(movies.duration >= 200) | (movies.genre == 'Drama')].head(10)
```

```python
(movies.duration >= 200) & (movies.genre == 'Drama')
```

```python
movies[(movies.genre == 'Crime')| (movies.genre == 'Adventure') |(movies.genre == 'Drama')]
```

```python
movies.genre.isin(['Crime', 'Adventure', 'Drama'])
```

```python
# short way using isin() method
movies[movies.genre.isin(['Crime', 'Adventure', 'Drama'])]
# same as movies[(movies.genre == 'Crime')| (movies.genre == 'Adventure') |(movies.genre == 'Drama')]
```
