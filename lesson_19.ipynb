```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head()
```

```python
# loc() selecting thing BY LABEL with argument[row, column]
ufo.loc[3:4, :]
```

```python
ufo.loc[2, :]
```

```python
ufo.loc[[0,2,6], :]
```

```python
ufo.loc[3:4, 'City']
```

```python
ufo.loc[3:4, ['City', 'State']]
```

```python
ufo.loc[3:4, 'City': 'State']
```

```python
ufo[ufo.City == 'Oakland']
```

```python
ufo.loc[ufo.City ==  'Oakland', :]
```

```python
# iloc() selecting thing BY INTEGER position.
# the range in argument will exclude the last row/column
ufo.iloc[:, [0,3]]
```

```python
ufo.iloc[2:20, [0,3]]
```

```python
ufo.iloc[2:20, 0:4]
```

```python
len(ufo.iloc[2:20, 0:4])
```

```python
ufo.iloc[2:20, 0:4].shape
```

```python
ufo[['City', 'State']]
```

```python
ufo.loc[:, ['City', 'State']]
```
