```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
pd.isnull(ufo).head()
```

```python
ufo.isnull().head()
```

```python
# loc[] is a LABEL based indexing, it includes both sides
ufo.loc[0:4,:]
```

```python
# iloc[] is a POSITION based indexing, it includes the LEFT but excludes the RIGHT
ufo.iloc[0:4,:]
```

```python
ufo.values
```

```python
# similar to Numpy slicing
ufo.values[0:4, :]
```

```python
# similar to range in Python
list(range(0,4))
```

```python
# loc[] LABEL indexing when selecting columns
ufo.loc[:, 'City':'State']
```

```python
# iloc[] POSITION indexing when selecting columns, it includes the RIGHT too
ufo.iloc[:, 0:4]
```

```python
# random sampling, produces different result every run
ufo.sample(n=3)
```

```python
# parameter random_state will give same result during every run
ufo.sample(n=3, random_state=40)
```

```python
# instead of specifying no of sample, can specify % of total row as frac=
ufo.sample(frac=0.75, random_state=88)
```

```python
# assuming 75% of the rows are training set 
train = ufo.sample(frac=0.75, random_state=88)
```

```python
# to put the remaining 25% of rows as test set, use below codes to avoid duplication
test = ufo.loc[~ufo.index.isin(train.index), :]
```
